<span style="color:#E50914; font-weight:bold; font-size:24px">About Dataset</span>

Netflix stands as a leading force in the realm of media and video streaming. With a staggering array of over 8,000 movies and TV shows accessible on their platform, as of mid-2021, their global subscriber count exceeds 200 million. This tabulated dataset comprehensively catalogues all offerings on Netflix, including vital details such as cast, directors, ratings, release year, duration, and more.

<span style="color:#E50914; font-weight:bold; font-size:20px">Data Overview</span>

The Netflix Titles dataset is a comprehensive compilation of movies and TV shows available on Netflix, covering various aspects such as the title type, director, cast, country of production, release year, rating, duration, genres (listed in), and a brief description. This dataset is instrumental for analyzing trends in Netflix content, understanding genre popularity, and examining the distribution of content across different regions and time periods.

<span style="color:#E50914; font-weight:bold; font-size:20px">Key Details:</span>

<span style="color:#E50914; font-weight: bold">Total Entries</span>: The dataset contains 8,809 entries, each representing a unique movie or TV show.

<span style="background-color:#800020; color:white;">Columns:</span>There are 12 columns in the dataset:

`show_id:` A unique identifier for each title.

`type:` The category of the title, which is either 'Movie' or 'TV Show'.

`title:` The name of the movie or TV show.

`director:` The director(s) of the movie or TV show. (Contains null values for some entries, especially TV shows where this information might not be applicable.)

`cast:` The list of main actors/actresses in the title. (Some entries might not have this information.)

`country:` The country or countries where the movie or TV show was produced.

`date_added:` The date the title was added to Netflix.

`release_year:` The year the movie or TV show was originally released.

`rating:` The age rating of the title.

`duration:` The duration of the title, in minutes for movies and seasons for TV shows.

`listed_in:` The genres the title falls under.

`description:` A brief summary of the title.

<span style="color:#E50914; font-weight:bold; font-size:20px">Importing libraries</span>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

<span style="color:#E50914; font-weight:bold; font-size:20px">Load dataset</span>

```python
df = pd.read_csv('netflix_titles.csv', encoding='unicode_escape')
```

```python
df.head()
```

```python
df =  df[df.columns[:12]]
df.head()
```

```python
df.shape
```

```python
df.info()
```

<span style="color:#E50914; font-weight:bold; font-size:20px">Imputation of missing values</span>

```python
df.isnull().sum().sort_values(ascending=False)
```

<span style="color:#E50914; font-weight:bold; font-size:20px">Replacing the null values of country with most common country in the data</span>

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

<span style="color:#E50914; font-weight:bold; font-size:20px">Replacing the null values of Cast and director by "Missing" and deleting all other rows which have missing values in them</span>

```python
df['director'].replace(np.nan, 'Missing', inplace= True)
df['cast'].replace(np.nan, 'Missing', inplace= True)

#delete other missing values
df = df.dropna()
```

<span style="color:#E50914; font-weight:bold; font-size:20px">Now we don't have any missing value in data</span>

```python
df.isnull().sum()
```

<span style="color:green; font-weight:bold; font-size:20px">How many movies or TV shows the top directors have directed?</span>

```python
# Group by director and type, and count occurrences
director_counts = df.groupby(['director', 'type']).size().unstack(fill_value=0)

# Sum the counts of movies and TV shows for each director
director_counts['Total'] = director_counts.sum(axis=1)

# Select the top 5 directors based on the total count
top_directors = director_counts.sort_values(by='Total', ascending=False).head(11)
top_directors =  top_directors.iloc[1:]

# Display the top 5 directors
print(top_directors)
```

```python
# Plotting
color = ['#E50914', 'green']
top_directors.drop(columns='Total').plot(kind='bar', color =color, stacked=True, figsize=(10, 6))
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Top 10 Directors by Total Count of Movies and TV Shows')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Type')
plt.show()
```

<span style="background-color:#800020; color:white;">Result: The top 10 directors have only directed movies except for Marcus Raboy who has also directed 1 TV show. </span>

<span style="color:#E50914; font-weight:bold; font-size:20px">Now Exploring the date added and release year column </span>

```python
# Convert the "date_added" column to string type
df["date_added"] = df["date_added"].astype(str)

# converting the date added column in the uniform format
df["date_added"] = df["date_added"].str.strip()

# convert dtype to datetime 
df["date_added"] = pd.to_datetime(df['date_added'])
# extract month and year
df['year_added'] = df['date_added'].dt.year
```

```python
df.head()
```

```python
release_year_count = df['release_year'].value_counts(ascending=False)
release_year_count = release_year_count.head(5)
release_year_count
```

```python
# plot the bar chart between release year and count
release_year_count.plot(kind = 'bar', color=color)
plt.xlabel('release year')
plt.ylabel('count')
plt.title('count of content released in top 5 year')

# Adding value labels on top of each bar
for i, count in enumerate(release_year_count):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')
    
plt.show()
```

<span style="background-color:#800020; color:white;">Result: The content that was released in 2018 is the most on netflix </span>

<span style="color:#E50914; font-weight:bold; font-size:20px">In which year most of the content was added on netflix? </span>

```python
# get the top five years in which content was added
year_added_count = df['year_added'].value_counts(ascending=False)
year_added_count =  year_added_count.head(5)
year_added_count
```

```python
# plot the bar chart between year in which content was added and count
year_added_count.plot(kind = 'bar', color=color)
plt.xlabel('year in which content was added')
plt.ylabel('count of content')
plt.title('count of content added on netflix in top 5 year')

# Adding value labels on top of each bar
for i, count in enumerate(year_added_count):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')
    
plt.show()
```

<span style="background-color:#800020; color:white;">Result: Most content was added on netflix in the year of 2019 </span>

<span style="color:#E50914; font-weight:bold; font-size:20px">What type of content was added on netflix in top 5 years? </span>

```python
contenttype_yearadded = df.groupby(['year_added', 'type']).size().unstack(fill_value=0)
contenttype_yearadded['total'] = contenttype_yearadded.sum(axis = 1)
contenttype_yearadded = contenttype_yearadded.sort_values(by='total', ascending=False)
contenttype_yearadded = contenttype_yearadded.head(5)
contenttype_yearadded
```

```python
# drop the total column now
contenttype_yearadded = contenttype_yearadded.drop(columns='total')
color = ['#E50914','green']
# plotting the chart between year added and type of content
plt.figure(figsize = (10,6))
contenttype_yearadded.plot(kind= 'bar', color = color)
plt.xlabel('year in which content was added')
plt.ylabel('count of content')
plt.title('count of content added on netflix in top 5 year')
plt.legend(loc="upper right")
plt.show()
```

```python
# Get the top 5 years based on the sum of counts across all content types
top_years = contenttype_yearadded.sum(axis=1).nlargest(5).index

# Calculate the number of rows and columns for subplots
num_rows = 1
num_cols = len(top_years)

# Plotting  pie charts for each of the top 5 years
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5))

for i, year in enumerate(top_years):
    yearly_counts = contenttype_yearadded.loc[year]
    axes[i].pie(yearly_counts, labels=yearly_counts.index, colors=color, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f'{year}')
    axes[i].axis('equal') 
plt.suptitle('Distribution of Content Types for Top 5 Years', y=1.05)
plt.tight_layout()
plt.show()
```

<span style="background-color:#800020; color:white;">Result: Movies were added more that TV shows on Neflix in top 5 years</span>

<span style="color:#E50914; font-weight:bold; font-size:20px">Now let's Explore the rating column </span>

```python
rating_count = df['rating'].value_counts(ascending=False)
rating_count =  rating_count.head(5)
rating_count
```

```python
# bar chart
color = ['red', 'green', 'maroon', 'purple', 'black']
rating_count.plot(kind = 'bar', color=color)
plt.xlabel('rating')
plt.ylabel('count of content')
plt.title('Top ratings')
# add the labels on each bar
for i, count in enumerate(rating_count):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')
plt.show()
```

<span style="background-color:#800020; color:white;">Result: The top ratings of netflix suggest most of the content is not suitable for children to watch.</span>

```python
# content type and their rating
rating_contenttype = df.groupby(['rating', 'type']).size().unstack(fill_value=0)
rating_contenttype['total'] = rating_contenttype.sum(axis = 1)
rating_contenttype =  rating_contenttype.sort_values(by='total', ascending=False)
rating_contenttype = rating_contenttype.head(5)
rating_contenttype
```

```python
# drop the total column now
rating_contenttype = rating_contenttype.drop(columns='total')
color = ['#E50914','green']
# plotting the bar chart between year added and type of content
plt.figure(figsize = (10,6))
rating_contenttype.plot(kind= 'bar', color = color)
plt.xlabel('ratings of types of content')
plt.ylabel('count')
plt.show()
```

```python

# Get the top 5 ratings based on the sum of counts across all content types
top_ratings = rating_contenttype.sum(axis=1).nlargest(5).index

# Calculate the number of rows and columns for subplots
num_rows = 1
num_cols = len(top_ratings)

# Plotting pie charts for each of the top 5 ratings
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5))

for i, rating in enumerate(top_ratings):
    content_counts = rating_contenttype.loc[rating]
    axes[i].pie(content_counts, labels=content_counts.index, colors=color, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f'{rating}')
    axes[i].axis('equal') 

plt.suptitle('Distribution of Content Types for Top 5 Ratings', y=1.05)
plt.tight_layout()
plt.show()
```
